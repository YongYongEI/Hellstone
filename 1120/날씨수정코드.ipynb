```python
import requests
from datetime import datetime, timedelta

# OpenWeatherMap API 키
API_KEY = '09806c8ab9fb92886b5a722c6359df79'  # OpenWeatherMap API 키를 입력하세요.
# 주요 동과 위도, 경도 정의
districts = {
    "동구": {
        "대동": (36.3241, 127.4349)
    },
    "중구": {
        "대흥동": (36.317, 127.418),
    },
    "서구": {
        "둔산동": (36.3528, 127.3823),
    },
    "유성구": {
        "노은동": (36.3832, 127.3308)
    },
    "대덕구": {
        "오정동": (36.3617, 127.4199)
    }
}

# 기상 상태 매핑
weather_mapping = {
    "clear sky": "맑음",
    "few clouds": "흐림",
    "scattered clouds": "흐림",
    "broken clouds": "흐림",
    "overcast clouds": "흐림",
    "rain": "비",
    "light rain": "비",
    "moderate rain": "비",
    "heavy rain": "비",
    "shower rain": "비",
    "snow": "눈",
    "light snow": "눈",
    "heavy snow": "눈",
    "fog": "안개",
    "mist": "안개",
    "haze": "안개"
}

default_weather = "기타"

# 내일 날짜 계산
def get_tomorrow_date():
    today = datetime.now()
    tomorrow = today + timedelta(days=1)
    return tomorrow.strftime('%Y-%m-%d')

# OpenWeatherMap description을 매핑
def map_weather_description(description):
    return weather_mapping.get(description, default_weather)

# 날씨 데이터 가져오기 (9시 데이터만)
def get_weather(lat, lon):
    url = f"http://api.openweathermap.org/data/2.5/forecast"
    params = {
        "lat": lat,
        "lon": lon,
        "appid": API_KEY,
        "units": "metric",  # 섭씨 사용
        "lang": "en"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        tomorrow_date = get_tomorrow_date()
        for item in data['list']:
            # 내일 날짜와 09:00:00에 해당하는 데이터 필터링
            if item['dt_txt'].startswith(tomorrow_date) and item['dt_txt'].endswith("09:00:00"):
                mapped_weather = map_weather_description(item['weather'][0]['description'])
                if mapped_weather in ["맑음", "흐림", "비", "눈", "기타", "안개"]:
                    return {
                        "시간": item['dt_txt'],
                        "날씨": mapped_weather,
                        "온도": item['main']['temp']
                    }
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

# 주요 동별 날씨 데이터 출력
for district, locations in districts.items():
    print(f"=== {district} ===")
    for dong, (lat, lon) in locations.items():
        print(f"동: {dong}")
        weather_data = get_weather(lat, lon)
        if weather_data:
            print(weather_data)
        else:
            print("날씨 데이터를 가져오지 못했습니다.")
        print()
```
